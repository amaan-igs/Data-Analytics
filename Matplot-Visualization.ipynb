# Pandas For Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./Datasets/Salesrec.csv') 
```

```python
# df.to_csv('updated-emp.csv', index=False)
```

```python
cbr = df.Region.value_counts()
cbr  #cbr is count by region
x = cbr.index
y = cbr.values
```

```python
sbr = df.groupby('Region')['Total Revenue'].sum() # sales by region
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summary')
plt.xlabel('Region')
plt.ylabel('Count')
ax =cbr.plot(x=x, y=y, kind='line', legend=False, style='r--')

plt.xticks(rotation=5, ha='center') 
```

```python
plt.figure(figsize = (15,5))
plt.title('Region wise Summary')
plt.xlabel('Region')
plt.ylabel('Count')
ax =cbr.plot(x=x, y=y, kind='line', legend=False, style='r--')

plt.xticks(rotation=5, ha='center') 

# add data labels

for i,j in enumerate(y):
    ax.text( i , j , j , ha='left' , va='bottom')

plt.plot(x , y , marker = 'o' , c = 'blue')
```

```python
plt.figure(figsize = (15,5)) 
plt.title("Region wise Summery") 
plt.xlabel("Region") 
plt.ylabel("Total Revenue")
           
ax = sbr.plot(x=sbr.index, y=sbr.values, kind="bar", legend=False)
plt.xticks(rotation=5, ha='center');
# Add data labels using text() 
for i, j in enumerate(sbr.values): 
    plt.text(i, j, format(j, ',.1f'), color='black', fontweight='bold', ha='center', va='bottom')
```
