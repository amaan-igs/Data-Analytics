# Pandas For Data Analysis

```python
import pandas as pd
```

```python
df = pd.read_csv('./Datasets/Salesrec.csv') 
```

```python
# df.to_csv('updated-emp.csv', index=False)
```

```python
df
```

```python
print(df.columns)
```

```python
cbr = df.Region.value_counts()
cbr  #cbr is count by region
```

```python
x = cbr.index
y = cbr.values
```

```python
cbr.plot.line(x=x, y=y)
```

```python
cbr.plot(x=x, y=y, kind='bar')
```

```python
cbr.plot(x=x, y=y, kind='pie')
```

```python
cbr.plot( x=x , y=y , kind='bar' , figsize=(12,6) , title ='Sales by Region' , 
         xlabel='Region' , ylabel='Sales' , grid=True , rot=70 , fontsize=12 , color='orange')
```

```python
cbr.plot(kind='pie', figsize=(6, 6), autopct='%1.1f%%', startangle=90, shadow=True, legend=True,
         labels=x, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen'])
```

```python
sbr = df.groupby('Region')['Total Revenue'].sum()
sbr # sales by region
```

```python
sbr.plot(kind='pie', figsize=(6, 6), autopct='%1.1f%%', startangle=90, shadow=True, 
         labels=x, colors=['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen'])
```
